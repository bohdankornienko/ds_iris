```python
data = []
with open('data/iris.data', 'r') as csvfile:
    iris_reader = csv.reader(csvfile)
    for row in iris_reader:
        features = row[:4]
        if 0 == len(features):
            break
            
        features = [float(x) for x in features]
        label = row[4]
        
        item = {'vec': features, 'label': label}
        data.append(item)
        
data
```

```python
data[0]
```

```python
for i, item in enumerate(data):
    colors = matplotlib.cm.rainbow(np.linspace(0, 1, len(data[i]['vec'])))
    
    sample = data[i]['vec']
    xs = [i for p in range(len(sample))]

    plt.scatter(xs, sample, color=colors)

plt.show()
```
